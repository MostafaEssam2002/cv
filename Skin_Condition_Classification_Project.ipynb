```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from keras.regularizers import l2
from tensorflow.keras.applications import VGG16 #feature extractor & transfer learning
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import RMSprop #improve model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau #stop training when improvements stop , reduce learning rate when learn stop


class SkinConditionClassifier:
    def __init__(self, train_dir, test_dir, img_size=(64, 64)):
        self.train_dir = train_dir
        self.test_dir = test_dir
        self.img_size = img_size
        self.cnn_model = None
        self.transfer_model = None
        self.callbacks = [
            EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)
        ]
        # Data Augmentation
        self.train_datagen = ImageDataGenerator(
            rescale=1.0 / 255,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True
        )
        self.test_datagen = ImageDataGenerator(rescale=1.0 / 255)

        # Load Data
        self.training_set = self.train_datagen.flow_from_directory(
            self.train_dir,
            target_size=self.img_size,
            batch_size=32,
            shuffle=True,
            class_mode='categorical'
        )
        self.test_set = self.test_datagen.flow_from_directory(
            self.test_dir,
            target_size=self.img_size,
            batch_size=32,
            class_mode='categorical'
        )

    def build_cnn_model(self):
        classifier = Sequential()

        # Convolutional Block 1
        classifier.add(Conv2D(64, (3, 3), activation='relu', input_shape=(*self.img_size, 3), kernel_regularizer=l2(0.001)))
        classifier.add(Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)))
        classifier.add(MaxPooling2D(pool_size=(2, 2)))
        classifier.add(BatchNormalization())
        classifier.add(Dropout(0.3))

        # Convolutional Block 2
        classifier.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.001)))
        classifier.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.001)))
        classifier.add(MaxPooling2D(pool_size=(2, 2)))
        classifier.add(BatchNormalization())
        classifier.add(Dropout(0.4))

        # Convolutional Block 3
        classifier.add(Conv2D(256, (3, 3), activation='relu', kernel_regularizer=l2(0.001)))
        classifier.add(Conv2D(256, (3, 3), activation='relu', kernel_regularizer=l2(0.001)))
        classifier.add(MaxPooling2D(pool_size=(2, 2)))
        classifier.add(BatchNormalization())
        classifier.add(Dropout(0.5))

        # Flattening and Fully Connected Layers
        classifier.add(Flatten())
        classifier.add(Dense(units=128, activation='relu'))
        classifier.add(Dropout(0.5))
        classifier.add(Dense(units=64, activation='relu'))
        classifier.add(Dropout(0.5))
        classifier.add(Dense(units=2, activation='softmax'))

        classifier.compile(optimizer=RMSprop(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
        self.cnn_model = classifier

    def train_cnn(self, epochs=50):
        history = self.cnn_model.fit(
            self.training_set,
            steps_per_epoch=len(self.training_set),
            epochs=epochs,
            validation_data=self.test_set,
            validation_steps=len(self.test_set),
            callbacks=self.callbacks
        )
        self.plot_training_history(history)

    def build_transfer_learning_model(self):
        base_model = VGG16(weights='imagenet', include_top=False, input_shape=(*self.img_size, 3))
        for layer in base_model.layers:
            layer.trainable = False

        x = base_model.output
        x = Flatten()(x)
        x = Dense(256, activation='relu')(x)
        x = Dropout(0.5)(x)
        x = Dense(128, activation='relu')(x)
        x = Dropout(0.5)(x)
        predictions = Dense(2, activation='softmax')(x)

        model = Model(inputs=base_model.input, outputs=predictions)
        model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
        self.transfer_model = model

    def train_transfer_learning(self, epochs=20):
        history = self.transfer_model.fit(
            self.training_set,
            steps_per_epoch=len(self.training_set),
            epochs=epochs,
            validation_data=self.test_set,
            validation_steps=len(self.test_set),
            callbacks=self.callbacks
        )
        self.plot_training_history(history)

    def plot_training_history(self, history):
        plt.figure(figsize=(10, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Loss Curves')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.show()

    def evaluate_model(self, model):
        test_loss, test_accuracy = model.evaluate(self.test_set, steps=len(self.test_set))
        print(f"Test Accuracy: {test_accuracy}")

    def plot_confusion_matrix(self, model):
        self.test_set.reset()
        predictions = np.argmax(model.predict(self.test_set), axis=1)
        y_true = self.test_set.classes
        cm = confusion_matrix(y_true, predictions)
        cmd = ConfusionMatrixDisplay(cm, display_labels=list(self.test_set.class_indices.keys()))
        cmd.plot(cmap='Blues')
        plt.show()

    def extract_features_for_ml(self, model):
        feature_extractor = Sequential(model.layers[:-1])
        train_features = feature_extractor.predict(self.training_set)
        test_features = feature_extractor.predict(self.test_set)

        train_labels = self.training_set.classes
        test_labels = self.test_set.classes

        return train_features, train_labels, test_features, test_labels

    def train_svm(self, features, labels, test_features, test_labels):
        svm = SVC(kernel='poly', probability=True)
        svm.fit(features, labels)
        y_pred = svm.predict(test_features)
        accuracy = accuracy_score(test_labels, y_pred)
        print(f"SVM Accuracy: {accuracy}")
        print(classification_report(test_labels, y_pred, target_names=list(self.test_set.class_indices.keys())))

    def train_knn(self, features, labels, test_features, test_labels):
        knn = KNeighborsClassifier(n_neighbors=5)
        knn.fit(features, labels)
        y_pred = knn.predict(test_features)
        accuracy = accuracy_score(test_labels, y_pred)
        print(f"KNN Accuracy: {accuracy}")
        print(classification_report(test_labels, y_pred, target_names=list(self.test_set.class_indices.keys())))


# Example Usage
train_dir = '/content/train'
test_dir = '/content/test'
classifier = SkinConditionClassifier(train_dir, test_dir)

# CNN Training
classifier.build_cnn_model()
classifier.train_cnn()
classifier.evaluate_model(classifier.cnn_model)
classifier.plot_confusion_matrix(classifier.cnn_model)

# Transfer Learning Training
classifier.build_transfer_learning_model()
classifier.train_transfer_learning()
classifier.evaluate_model(classifier.transfer_model)
classifier.plot_confusion_matrix(classifier.transfer_model)

# Train ML Models (SVM and KNN)
train_features, train_labels, test_features, test_labels = classifier.extract_features_for_ml(classifier.cnn_model)
classifier.train_svm(train_features, train_labels, test_features, test_labels)
classifier.train_knn(train_features, train_labels, test_features, test_labels)

```
